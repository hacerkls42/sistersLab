```python
# Visualize birds wingspan
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.rcParams["figure.figsize"] = (12,6)
wingspan = birds.MaxWingspan 
wingspan.plot()
plt.show()
```

```python
# Check outliers
birds[birds["MaxWingspan"] > 500]
```

```python
# Label outliers
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, "o", color="steelblue")
    if birds['MaxWingspan'][i] > 500:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=10,)
    
plt.show()
```

```python
# 1- Remove outliers in plot

plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    if birds['Name'][i] not in ['Bald eagle', 'Prairie falcon']:
        plt.plot(x, y, "o", color="steelblue")
plt.show()
```

```python
# 2- Remove outliers before plot
birds = birds[birds["MaxWingspan"] < 500]
```

```python
birds.info()
```

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
wingspan = birds.MaxWingspan 
wingspan.plot(kind="bar")
plt.show()
```

```python
category_wingspan = birds.groupby("Category")["MaxWingspan"].mean().sort_values(ascending=False)
category_wingspan.head()

```

```python
category_wingspan = birds.groupby("Category")["MaxWingspan"].mean().sort_values(ascending=False)
category_wingspan.plot(kind="bar", figsize=(12,6), title="Average Max Wingspan by Category") #kategorilere göre ortalama kanat çırpışı 
plt.ylabel("Ortalama Max Wingspan (cm)")
plt.xlabel("Category")
plt.show()
```

```python

# 1. En ağır 10 kuşu seç
heaviest_birds = birds.sort_values(by="MaxBodyMass", ascending=False).head(10)

plt.barh( # 2. Yatay
    heaviest_birds["Name"],      # Kuş isimleri 
    heaviest_birds["MaxBodyMass"],# Ağırlık değerleri 
    color="darkred"
)

plt.xlabel("Weight (g)")         
plt.ylabel("Bird Species")       
plt.title("Top 10 Heaviest Birds")  # En Ağır 10 Kuş

plt.gca().invert_yaxis() # 4. En ağır kuş en üstte olsun 

plt.show()
```

```python
def bird_stats(df):
    stats = {
        "Total Birds": len(df),  # Toplam Kuş Sayısı
        "Longest Bird": df.loc[df["MaxLength"].idxmax(), "Name"],  # En uzun kuş
        "Max Wingspan": df["MaxWingspan"].max(),  # En geniş kanat açıklığı
        "Categories Count": df["Category"].nunique()  # Kategori sayısı
    }
    return pd.Series(stats, name="Bird Statistics")

# Fonksiyon çağrısında print yok, Jupyter hücresi doğrudan sonucu gösterecek
bird_stats(birds)
```
